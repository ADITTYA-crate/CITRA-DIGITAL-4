```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('patrick.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

m = 3
m_bit = np.floor(gray / (256 / (2**m))) * (256 / (2**m))
m_bit = m_bit.astype(np.uint8)

brightness = cv2.convertScaleAbs(img_rgb, alpha=1, beta=50)

contrast = cv2.convertScaleAbs(img_rgb, alpha=1.5, beta=0)

hist = cv2.calcHist([gray],[0],None,[256],[0,256])

equalized = cv2.equalizeHist(gray)

plt.figure(figsize=(12,10))

plt.subplot(3,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(3,3,2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.subplot(3,3,3)
plt.imshow(binary, cmap='gray')
plt.title("Binary")
plt.axis("off")

plt.subplot(3,3,4)
plt.imshow(m_bit, cmap='gray')
plt.title("M-bit (3 bit)")
plt.axis("off")

plt.subplot(3,3,5)
plt.imshow(brightness)
plt.title("Brightness")
plt.axis("off")

plt.subplot(3,3,6)
plt.imshow(contrast)
plt.title("Contrast")
plt.axis("off")

plt.subplot(3,3,7)
plt.plot(hist)
plt.title("Histogram")

plt.subplot(3,3,8)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalization")
plt.axis("off")

plt.tight_layout()
plt.show()
```
